```python
import pandas as pd, numpy as np

# 1. Load Data
c = pd.read_csv('customers.csv')
s = pd.read_csv('sales_transactions.csv')

# 2. Missing Values & Standardization (เก็บคะแนนข้อ 1 และ 3) # ลบขยะในเบอร์โทร และเติม 0 (Standardization)
c['age'] = c['age'].fillna(c['age'].median())
c['phone_number'] = c['phone_number'].fillna('0').astype(str).str.replace(r'[^0-9+]', '', regex=True)
# โจทย์สั่งเติม 0
s['promotion_id'] = s['promotion_id'].fillna(0) 

# 3. Data Type Conversion (เก็บคะแนนข้อ 2)
# ฟังก์ชันสั้นๆ สำหรับแปลงวันที่และสุ่มเวลา 9am - 5pm (32400 - 61200 วินาที)
fix_dt = lambda x: pd.to_datetime(x.fillna('2024-01-01'), errors='coerce', format='mixed').dt.normalize() + pd.to_timedelta(np.random.randint(32400, 61201, len(x)), unit='s')

c['join_date'] = fix_dt(c['join_date'])
c['last_purchase_date'] = fix_dt(c['last_purchase_date'])
s['date'] = fix_dt(s['date'])

# 4. Deliverables (ส่งไฟล์ตามชื่อที่โจทย์กำหนด)
c.to_csv('output_z/customers_cleaned.csv', index=False)
s.to_csv('output_z/sales_transactions_cleaned.csv', index=False)

print("✅ Done!")
```
