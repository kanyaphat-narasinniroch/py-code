```python
import pandas as pd, matplotlib.pyplot as plt ;
from matplotlib.backends.backend_pdf import PdfPages

df = pd.read_csv('output/sales_transactions_cleaned.csv').assign(date = lambda x: pd.to_datetime(x['date']).dt.strftime('%Y-%m'))
df ['revenue'] = (df['quantity']) * (df['price']) - pd.to_numeric(df['discount_amount'], errors='coerce').fillna(0)

m_stats = df.groupby('date').agg( rev = ('revenue','sum'), tx = ('transaction_id', 'nunique')).assign(aov = lambda x: x['rev'] / x['tx']).sort_index().reset_index()
top3 = m_stats.nlargest(3, 'rev') [['date', 'rev']].assign(rev = lambda x: x['rev'].apply(lambda v: f"${v:,.2f}" ))
# print(top3)
with PdfPages('output/Session1_SalesTrends.pdf') as pdf:
    fig, ax = plt.subplots(3, 1, figsize = (10, 12)); fig.tight_layout(pad = 5.0)
    for i, (col, c, title) in enumerate ([('rev', 'red', 'Total Revenue'), ('tx', 'blue', 'Transactions'), ('aov', 'orange', 'Average order value')]):
        ax[i].plot(m_stats['date'], m_stats[col], marker = 'o', color = c); ax[i].set_title(title, weight = 'bold'); ax[i].tick_params(axis = 'x', rotation = 40); ax[i].grid(ls = ':')
    pdf.savefig(fig, bbox_inches = 'tight');
    plt.close()

    fig, ax = plt.subplots(figsize = (6, 2)); ax.axis('off'); tbl = ax.table(cellText = top3.values, colLabels = ['Month', 'Total Revenue'], loc = 'center', cellLoc = 'center'); tbl.scale(1, 2); ax.set_title('Top 3 Months', weight = 'bold')
    pdf.savefig(fig, bbox_inches = 'tight');
    plt.close()
```
