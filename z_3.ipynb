```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# 1. Data Prep (เหมือนเดิม)
df = pd.read_csv('output_z/sales_transactions_cleaned.csv')
df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m')
df['rev'] = df.quantity * df.price - pd.to_numeric(df.discount_amount, errors='coerce').fillna(0)

m = df.groupby('date').agg(rev=('rev','sum'), tx=('transaction_id','nunique')).reset_index()
m['aov'] = m.rev / m.tx
t3 = m.nlargest(3, 'rev').assign(rev=lambda x: x.rev.map('${:,.2f}'.format))

# 2. PDF & Plots
with PdfPages('output_z/Session1_SalesTrends.pdf') as pdf:
    for col, color, title in [('rev','red','Total Revenue'), ('tx','blue','Transactions'), ('aov','green','Average order value')]:
        fig, ax = plt.subplots(figsize=(12, 5))
        ax.plot(m['date'], m[col], marker='o', color=color)
        ax.set_title(title, weight='bold')
        ax.tick_params(axis='x', rotation=45)
        ax.grid(ls='--', alpha=0.5) # <--- 1. เพิ่มเส้นตาราง (Grid)
        pdf.savefig(fig, bbox_inches='tight'); plt.close()

    # 3. Table Page
    fig, ax = plt.subplots(figsize=(12, 2)); ax.axis('off')
    tbl = ax.table(cellText=t3[['date', 'rev']].values, colLabels=['Month', 'Rev'], loc='center', cellLoc='center')
    tbl.scale(1, 2)
    
    # <--- 2. เพิ่มสีและตัวหนาที่หัวข้อ (Loop สั้นๆ) ---
    for c in range(2): 
        tbl[0, c].set_facecolor('#2A7AE0') # สีพื้นหลังหัวข้อ
        tbl[0, c].set_text_props(color='white', weight='bold') # ตัวหนา + สีขาว
        
    pdf.savefig(fig, bbox_inches='tight'); plt.close()

print('✅ Done')
```
